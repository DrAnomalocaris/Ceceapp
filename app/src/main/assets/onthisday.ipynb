```python
import re
banned_words=[
    "suicide",
    "murder",
    "kill",
    "sexual",
    "assault",
    "died",
    "shoot",
    "executed"
]
cache2={}
n=0
removed=0
for (d,m),c in cache.items():
    if not m in cache2:
        cache2[m] = {}
    if not d in cache2[m]:
        cache2[m][d] = []
    for e in c:
        t= e['text']
        if "[" in t:
            t= re.sub(r'\[.*?\]', '', t).strip().replace("[","").replace(']','').replace(" , ",", ").replace("  "," ")
        if "(" in t:
            t= re.sub(r'\(.*?\)', '', t).strip().replace("(","").replace(')','').replace(" , ",", ").replace("  "," ")
        if any([word.lower() in t.lower() for word in banned_words]):
            removed+=1
            continue

        cache2[m][d].append({"year":e['year'],"text":t})
        n+=1
print(f"total entries: {n}, removed: {removed}")            
    
```

```python
import json
jsonstr = f"const onThisDay = {json.dumps(cache2)} ;"
jsonstr = jsonstr.replace('= {"',   '=\n{"')
jsonstr = jsonstr.replace('": {',   '":{\n    ')
jsonstr = jsonstr.replace('"}, {"', '"},\n        {"')
jsonstr = jsonstr.replace('], "',   '],\n    "')
jsonstr = jsonstr.replace('}]}, "', '}]},\n"')

print(jsonstr)
```
